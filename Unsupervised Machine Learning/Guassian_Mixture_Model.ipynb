# Gaussian Mixture Model (GMM)

## ðŸ“Œ Introduction
A **Gaussian Mixture Model (GMM)** is a probabilistic model that assumes data is generated from a mixture of several Gaussian distributions with unknown parameters.  
It is widely used in clustering, density estimation, and pattern recognition.

---

## ðŸ”¹ Key Concepts
- **Gaussian Distribution**: A bell-shaped curve defined by mean (Î¼) and variance (ÏƒÂ²).
- **Mixture Model**: A combination of multiple probability distributions.
- **Latent Variable**: Each data point is assumed to belong to one of the Gaussian components (hidden variable).

---

## ðŸ”¹ Working of GMM
1. Assume `k` Gaussian components in the data.
2. Initialize parameters: mean (Î¼), covariance (Î£), and mixing coefficient (Ï€).
3. Use **Expectation-Maximization (EM) algorithm**:
   - **E-step**: Estimate the probability that each data point belongs to a Gaussian component.
   - **M-step**: Update Gaussian parameters (Î¼, Î£, Ï€) to maximize likelihood.
4. Repeat until convergence.

---

## ðŸ”¹ Advantages
- Can model **complex, non-linear distributions**.
- Provides **soft clustering** (probability of belonging to each cluster).
- More flexible than K-Means.

## ðŸ”¹ Limitations
- Requires specifying number of components `k`.
- Can converge to **local optima**.
- Computationally expensive for large datasets.

---

## ðŸ”¹ Applications
- Image segmentation  
- Speech recognition  
- Anomaly detection  
- Customer segmentation  

---


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

# 1. Generate sample dataset
X, y_true = make_blobs(n_samples=5000, centers=10, cluster_std=0.6, random_state=42)

# 2. Fit Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
labels = gmm.predict(X)

# 3. Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("Gaussian Mixture Model Clustering")
plt.show()

# 4. Evaluation Metrics
log_likelihood = gmm.score(X) * len(X)   # total log-likelihood
aic = gmm.aic(X)
bic = gmm.bic(X)

# For silhouette, need at least 2 clusters
if len(set(labels)) > 1:
    silhouette = silhouette_score(X, labels)
else:
    silhouette = "Not applicable (only one cluster found)"

print("âœ… Evaluation Metrics for GMM")
print(f"Log Likelihood: {log_likelihood:.4f}")
print(f"AIC: {aic:.4f}")
print(f"BIC: {bic:.4f}")
print(f"Silhouette Score: {silhouette}")
```
